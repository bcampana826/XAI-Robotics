```python
import shap
import numpy as np
from sklearn.linear_model import LinearRegression
from lime import lime_tabular
from stable_baselines3 import PPO
import matplotlib.pyplot as plt
import gymnasium as gym

shap.initjs()
```

```python
env = gym.make("Pendulum-v1")
```

```python
model = PPO("MlpPolicy", env,verbose=10000,  tensorboard_log="./PPO_PEND_TENSORBOARD/")

#model.learn(total_timesteps=2000000)

# Save the trained model
#model.save("ppo_pend.zip")
model = PPO.load("resources/ppo_pend.zip",env,verbose=100)
```

```python
observation, info = env.reset()

XTrain = []
YTrain = []
done = False

for i in range(10000):
    ter = done = False
    x=0
    while not done and not ter and x < 200:
        action = model.predict(observation,deterministic=False)
        XTrain.append(observation)
        YTrain.append(action[0])
    
        observation, reward, done, ter, _ = env.step(action[0])
        x+=1

print(len(XTrain))
```

```python
XTest = []
YTest = []

observation, _ = env.reset()
ter = done = False
x=0
while not done and not ter and x < 200:
    action = model.predict(observation,deterministic=False)
    XTest.append(observation)
    YTest.append(action[0])

    observation, reward, done, ter, _ = env.step(action[0])
    x+=1
```

```python
def f(X):
    return model.predict(X,deterministic=True)[0]
```

```python
XTrain_np = np.array(XTrain)
XTest_np = np.array(XTest)


explainer = shap.KernelExplainer(f, shap.sample(XTrain_np, 200))

shap_values = explainer.shap_values(XTest_np)
print("shap_values =", shap_values)
print("base value =", explainer.expected_value)
```

```python
shap.force_plot(explainer.expected_value[0], shap_values[0], XTest_np, feature_names=["x_y","theta","torque"])
```

```python
shap.decision_plot(explainer.expected_value[0],shap_values[0], feature_names=["x_y","theta","torque"])
```

```python
print(f'Prediction for 1st sample in X_test: {f(XTest_np[0])[0]}')

shap.force_plot(explainer.expected_value[0], shap_values[0][0,:], XTest_np[0],link="logit")
```

```python
shap.summary_plot(shap_values[0], feature_names=["x_y","theta","torque"])
```

```python
YTest_np= np.array(YTest)

plt.plot(YTest_np)
plt.show()
```
